## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#Add numpy to run code for pyplot x-axis
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Merge data
combined_df = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")
combined_df.head()
```

```python
# Count mice via Mouse ID
mouse_count = combined_df["Mouse ID"].count()
mouse_count
```

```python
# Obtain duplicated Mouse IDs, Timepoints
duplicated_rows = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicated_rows
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
all_duplicate_rows = combined_df[combined_df.duplicated(["Mouse ID"])]
all_duplicate_rows
```

```python
# Drop duplicate mice (i.e., Mouse ID)
drop_dup_mice_df = combined_df.drop_duplicates(["Mouse ID"])
drop_dup_mice_df
```

```python
# Count duplicate mice
drop_dup_mice_count = drop_dup_mice_df["Mouse ID"].count()
drop_dup_mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
mean = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()

median = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()

variance = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()

standard_dv = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()

sem = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_stat_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dv,
"SEM": sem})

summary_stat_df.round(2)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Sort by "Drug Regimen", "Mouse ID" and "Timepoint"
sort_values_df = combined_df.sort_values(["Drug Regimen", "Mouse ID", "Timepoint"], ascending=True)
sort_values_df
```

```python
# Identify max Timepoint
max_df = combined_df.groupby('Drug Regimen')['Timepoint'].max()
max_df
```

```python
# Sort by max Timepoint
max_time_df = sort_values_df.loc[sort_values_df["Timepoint"] == 45]
max_time_df.head()
```

```python
# Regimen: Capomulin
# Select only data from Capomulin
capomulin_df = max_time_df[max_time_df["Drug Regimen"].isin(["Capomulin"])]
capomulin_df.head()
```

```python
#Sort values in ascending order by "Tumor Volume (mm3)"
capomulin_obj = capomulin_df.sort_values(["Tumor Volume (mm3)"], ascending=True)
capomulin_obj.head()
```

```python
# Generate Tumor Volume Dataframe
capomulin_obj = capomulin_obj["Tumor Volume (mm3)"]
capomulin_obj.head().round(2)
```

```python
#IQR Evaluation - Capomulin

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

quartiles = capomulin_obj.quantile([.25,.5,.75]).round(2)
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = (upperq - lowerq).round(2)
print(f"Capomulin Regimen: Lower Quartile: {lowerq}")
print(f"Capomulin Regimen: Upper Quartile: {upperq}")
print(f"Capomulin Regimen: Interquartile Range: {iqr}")
print(f"Capomulin Regimen: Median: {quartiles[0.5]}")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Capomulin Potential Outliers: Values Below {lower_bound}")
print(f"Capomulin Potential Outliers: Values Above {upper_bound}")
```

```python
# Capomulin Regimen: Box Plot
#Generate a box plot of the final tumor volume of each mouse across this regimen
plot, axis_1 = plt.subplots()
axis_1.set_title("Capomulin Regimen: Final Tumor Volume")
axis_1.set_ylabel("Final Tumor Volume (mm3)")
axis_1.boxplot(capomulin_obj)
plt.show()    
```

```python
# Regimen: Ramicane

# Select only data from Ramicane
ramicane_df = max_time_df[max_time_df["Drug Regimen"].isin(["Ramicane"])]

#Sort values in ascending order by "Tumor Volume (mm3)"
ramicane_obj = ramicane_df.sort_values(["Tumor Volume (mm3)"], ascending=True)

# Generate Tumor Volume Dataframe
ramicane_obj = ramicane_obj["Tumor Volume (mm3)"]

#IQR Evaluation - Ramicane

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

quartiles = ramicane_obj.quantile([.25,.5,.75]).round(2)
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = (upperq - lowerq).round(2)
print(f"Ramicane Regimen: Lower Quartile: {lowerq}")
print(f"Ramicane Regimen: Upper Quartile: {upperq}")
print(f"Ramicane Regimen: Interquartile Range: {iqr}")
print(f"Ramicane Regimen: Median: {quartiles[0.5]}")
lower_bound = lowerq - (1.5*iqr).round(2)
upper_bound = upperq + (1.5*iqr).round(2)
print(f"Ramicane Potential Outliers: Values Below {lower_bound}")
print(f"Ramicane Potential Outliers: Values Above {upper_bound}")
```
